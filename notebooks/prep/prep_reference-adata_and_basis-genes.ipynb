# Prepare Visium-OCT data

```python
adata_vis = sc.read_h5ad('/nfs/team205/heart/anndata_objects/Foetal/visium_foetal_healthy-all_raw.h5ad')
adata_vis.var_names_make_unique()

# subset OCT data
adata_vis = adata_vis[adata_vis.obs['modality']=='Visium']

# filter genes
sc.pp.filter_genes(adata_vis, min_cells=3)

adata_vis
```

```python
# remove "Adventitia - Coronary" from the annotation
# At the 1st round of cross-validatiaon, "Adventitia - Coronary" showed poor prediction accuracy (f1 score < 0.5) compared to other labels.
# Therefore, will remove from the label list and re-try cross validataion
adata_vis.obs['annotation_final_mod'] = adata_vis.obs['annotation_final_mod'].astype('str')
mask = adata_vis.obs['annotation_final_mod']=='Adventitia - Coronary'
adata_vis.obs.loc[adata_vis.obs_names[mask],'annotation_final_mod']='unassigned'
adata_vis.obs['annotation_final_mod'].value_counts()
```

```python
# tidy up
obs_columns = ['array_row', 'array_col', 'sample','donor', 'donor_type', 'region', 'region_finest', 'age','week', 
               'gender','modality','sample_for_cell2loc','annotation_final_mod']
adata_vis.obs = adata_vis.obs[obs_columns]
uns_keys = list(adata_vis.uns.keys())
for key in uns_keys:
    del adata_vis.uns[key]
obsp_keys = list(adata_vis.obsp.keys())
for key in obsp_keys:
    del adata_vis.obsp[key]
obsm_keys = list(adata_vis.obsm.keys())
for key in obsm_keys:
    if key!='spatial':
        del adata_vis.obsm[key]
adata_vis
```

```python
# save
adata_vis.write('/nfs/team205/kk18/github/TissueTypist/tissuetypist/visiumsd_oct_raw.h5ad')
```

# Prepare gene features for the training

### Log-normalise

```python
# log-normalise
sc.pp.normalize_total(adata_vis, target_sum=1e4)
sc.pp.log1p(adata_vis)
adata_vis
```

### Highly-variable genes

```python
# hvg with subsampling
ad = adata_vis[adata_vis.obs['annotation_final_mod']!='unassigned']
ad = sctk.subsample(ad,fraction=1,groupby='annotation_final_mod',max_n=1000)
print(ad.obs['annotation_final_mod'].value_counts())

sc.pp.highly_variable_genes(ad, min_disp=0.5, min_mean=0.0125, max_mean=3, batch_key='sample_for_cell2loc')
hvgs_sub = list(ad.var_names[ad.var['highly_variable']])
print(len(hvgs_sub))
sc.pl.highly_variable_genes(ad)
```

### DEGs

```python
logfoldchanges_thresh = 1
group_mean_thresh = 0.5
```

```python
ad = adata_vis[adata_vis.obs['annotation_final_mod']!='unassigned']
deg_df = tt.rank_genes_groups_with_mean(ad,groupby='annotation_final_mod',log2fc_min=0)
```

```python
mask1 = deg_df['logfoldchanges']>=logfoldchanges_thresh
mask2 = deg_df['group_mean']>=group_mean_thresh
degs = list(set(deg_df[mask1&mask2]['names']))
print(len(degs))
deg_df[mask1&mask2]['group'].value_counts()
```

```python
from matplotlib_venn import venn2
venn2(subsets=(set(hvgs_sub),set(degs)),set_labels=('hvgs','degs'))
```
